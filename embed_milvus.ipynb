```python
load_dotenv()

folder_id = os.getenv("GD_FOLDER_ID") 
milvus_uri = os.getenv("MILVUS_URI")

local_folder = "./data"
collection_name = "candidate_colpali"

gdm = GD2MilvusManager(
    milvus_uri=milvus_uri,
    collection_name=collection_name,
    login=False, # True if not wanna auto login
    # device='mps'scrip
)

mvc = gdm.milvus

# authenticate gdrive account
# gdm.authenticate() 
```

```python
gdm.download_files_recursive(folder_id, local_folder, auto_embed=False) # set auto_embed to True to automatically embed on milvus
```

```python
from pathlib import Path

folder = Path(local_folder)
filepaths = [str(file) for file in folder.rglob("*.pdf") if file.is_file()]

mvc(drop=True)
for filepath in filepaths:
    category = os.path.basename(os.path.dirname(filepath))
    mvc.store_pdf_images_in_milvus(filepath, category)
```
